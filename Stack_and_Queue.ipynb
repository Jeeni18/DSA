# **Stack and Queue Data Structures**
This notebook demonstrates implementation of Stack and Queue from scratch using Python lists.
The aim is to understand LIFO and FIFO behavior and analyze their time complexity.


Stack follows LIFO (Last In First Out).
Operations:
- push → add element
- pop → remove top element
- peek → view top
Time Complexity: O(1)


**Stack Implementation**

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, value):
        self.items.append(value)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return "Stack Underflow"

    def peek(self):
        if not self.is_empty():
            return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        print(self.items)
```

**Stack Usage**

```python
s = Stack()
s.push(10)
s.push(20)
s.push(30)

s.display()
print("Popped:", s.pop())
print("Top Element:", s.peek())
```

# **Queue Theory**

Queue follows FIFO (First In First Out).
Operations:
- enqueue → add element
- dequeue → remove front
Time Complexity: O(1)


**Queue Implementation**

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, value):
        self.items.append(value)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        return "Queue Underflow"

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        print(self.items)
```

**Queue Usage**

```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)

q.display()
print("Dequeued:", q.dequeue())
q.display()
```

# **Complexity Discussion**

Stack:
push → O(1)
pop → O(1)

Queue (list-based):
enqueue → O(1)
dequeue → O(n) because elements shift

Using collections.deque can make dequeue O(1).
